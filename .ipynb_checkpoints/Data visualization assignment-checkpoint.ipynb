<b>In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.</b>

```python
# Importing the necessary packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
# Reading the dataset
iris = datasets.load_iris()
```

```python
# Reducing the dimensions for 3D
X_reduced = PCA(n_components = 3).fit_transform(iris.data)
Y = iris.target

#Visualising the 3D chart
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=134)

# Reorder the labels to have colors matching the cluster results
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=Y, cmap=plt.cm.rainbow_r, edgecolor='k')

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3 component PCA')

plt.show()
```
